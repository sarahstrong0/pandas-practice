```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = 1 
n_samples = 1000
n_simulations = 1000 

means = []

for _ in range(n_simulations):
    # generate 1000 exponentially distributed random samples
    samples = np.random.exponential(scale=tau, size=n_samples)
    # find the mean, repeated 1000 times
    means.append(np.mean(samples))

# create histogram of the means
plt.hist(means, bins=30, density=True, alpha=0.6, color='b')
plt.xlabel('Mean Lifetime')
plt.ylabel('Density')
plt.title('Histogram of Mean Lifetime of Qubits')
plt.show()
```

`density=True` normalizes the lifetime of the qubits, with the area under the curve = 1,making it represent a probability density rather than just the frequency of occurences. 

The distribution is Gaussian (normal distribution).

Central limit theorem states that the distribution of the sample means will appraoch a normal distribution for large sample sizes, regardless of the shape of the original distribution (so long as the original distribution has a finite mean and variance). 

Given that we have found the expectation value and variance to be finite in part A and B, Central Limit Theorem holds
